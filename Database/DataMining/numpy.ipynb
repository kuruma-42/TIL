```python
import numpy as np 
test1 = np.array([0,1])
test2 = np.array([0.4,1])
test3 = np.array([0,-1])
test4 = np.array([1,2])
```

```python
import matplotlib.pyplot as plt 

plt.scatter(test1[0], test1[1], label='A')
plt.scatter(test2[0], test2[1], label='B')
plt.scatter(test3[0], test3[1], label='C')
plt.scatter(test4[0], test4[1], label='D')
plt.legend()
# plt.axis([0,0, 2,2])
plt.show()
```

```python
### 코사인유사도 
from numpy.linalg import norm 
import numpy as np 

def cos_sim(A, B):
    return np.dot(A, B)/(norm(A) * norm(B))

print(cos_sim(test1, test2))
print(cos_sim(test1, test3))
print(cos_sim(test1, test4))
```

```python
# 유클리드 거리 
xa, ya = 1,2
xb, yb = 0,3
a = np.array((xa ,ya))
b = np.array((xb, yb))

dist = np.sqrt((xa-xb)^2 + (ya-yb)^2)
dist
```

```python
dist = np.linalg.norm(a-b)
dist
```

```python
from scipy.spatial import distance
a = (1, 2, 3)
b = (4, 5, 6)
dist = distance.euclidean(a, b)
dist
```
