```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()

x_train, x_test, y_train, y_test = train_test_split(cancer['data'],
                                                    cancer['target'],
                                                    stratify=cancer['target'], # 층화추출
                                                    random_state=0) # seed를 준다.
cancer['target_names']
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
import numpy as np
np.bincount(cancer['target']) # 갯수 카운트 2:3 정도의 비율
```

```python
# 디시젼 트리 모델 만들고 학습시켜보세요. 정확도를 출력해보세요.
# y 연속값 -> 회귀 Refression
#   범주형 -> 분류 Classification
from sklearn.tree import DecisionTreeClassifier # 분류기

model = DecisionTreeClassifier().fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test) # 정확도를 찍는게 .score
```

```python
# 특성중요도
model.feature_importances_
```

- max_depth 조정

```python
# max_depth를 3으로 조정해보세요.
from sklearn.tree import DecisionTreeClassifier # 트리가 계속 내려가는 성격이 있기 때문에 depth를 미리 넣어 준다.
model3 = DecisionTreeClassifier(max_depth=3, random_state=0).fit(x_train, y_train)
model3.score(x_train, y_train), model3.score(x_test, y_test)
```

```python
#max_depth를 5로 조정해보세요.
from sklearn.tree import DecisionTreeClassifier # 트리가 계속 내려가는 성격이 있기 때문에 depth를 미리 넣어 준다. 5로 하면 성능 떨어진다.
model5 = DecisionTreeClassifier(max_depth=5, random_state=0).fit(x_train, y_train)
model5.score(x_train, y_train), model5.score(x_test, y_test)
```

# Decision Tree - Regression Tree


```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
boston = load_boston()

x_train, x_test, y_train, y_test = train_test_split(boston['data'],
                                                    boston['target'],
                                                    random_state=0)
```

```python
# 디시젼트리 회귀모델로 모델을 만들고, 학습 시켜 보세요. 정확도를 출력하세요.
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(),fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```

```python
model.tree_.max_depth
```

```python
model.feature_importances_
```

```python
#특성 중요도 시각화
import pandas as pd
fi = pd.Series(model.feature_importances_, index=boston['feature_names'])
fi[fi!=0].sort_values(ascending=False).plot(kind='bar', figsize=(10, 5))
```

```python
# graphviz를 사용하여 시각화
from sklearn.tree import export_graphviz
export_graphviz(model, out_file='model.dot',
                feature_names=boston['feature_names'],
                impurity=True, filled=True)
```

```python
with open('model.dot') as f:
    data = f.read()
graphviz.Source(data)
```
